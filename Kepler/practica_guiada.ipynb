## Ejercicio 2: Clasificador de préstamos

```python
def clasificar_prestamo(dias):
    if dias <= 21:
        estado = "A tiempo"
        penalizacion = 0
    elif dias <= 30:
        estado = "Retraso leve"
        dias_extra = dias - 21
        penalizacion = dias_extra * 0.50
    else:
        estado = "Retraso grave"
        dias_extra = dias - 30
        penalizacion = dias_extra * 1.00

    return {
        "estado": estado,
        "penalizacion": penalizacion
    }


# Datos de prueba
prestamos_dias = [15, 22, 18, 35, 25, 12, 40, 19, 28, 33]

# Contadores
a_tiempo = 0
retraso_leve = 0
retraso_grave = 0
penalizacion_total = 0

# Procesar préstamos
for dias in prestamos_dias:
    resultado = clasificar_prestamo(dias)
    
    if resultado["estado"] == "A tiempo":
        a_tiempo += 1
    elif resultado["estado"] == "Retraso leve":
        retraso_leve += 1
    else:
        retraso_grave += 1
    
    penalizacion_total += resultado["penalizacion"]

# Mostrar estadísticas
print("Estadísticas de préstamos")
print("A tiempo:", a_tiempo)
print("Retraso leve:", retraso_leve)
print("Retraso grave:", retraso_grave)
print("Penalización total:", penalizacion_total, "€")
```

## Ejercicio 3: Análisis de categorías

```python
prestamos = [
    {"titulo": "1984", "categoria": "Ficción", "dias": 15},
    {"titulo": "Sapiens", "categoria": "Ensayo", "dias": 22},
    {"titulo": "Watchmen", "categoria": "Cómic", "dias": 18},
    {"titulo": "El Quijote", "categoria": "Ficción", "dias": 30},
    {"titulo": "Breve historia", "categoria": "Ensayo", "dias": 25},
    {"titulo": "Batman", "categoria": "Cómic", "dias": 12}
]

# Agrupar préstamos por categoría
categorias = {}

for prestamo in prestamos:
    categoria = prestamo["categoria"]
    dias = prestamo["dias"]

    if categoria not in categorias:
        categorias[categoria] = []

    categorias[categoria].append(dias)

# Calcular promedio de días por categoría
print("Promedio de días por categoría:")
for categoria, lista_dias in categorias.items():
    promedio = sum(lista_dias) / len(lista_dias)
    print(categoria, ":", promedio)

# Identificar la categoría más prestada
categoria_mas_prestada = None
max_prestamos = 0

for categoria, lista_dias in categorias.items():
    if len(lista_dias) > max_prestamos:
        max_prestamos = len(lista_dias)
        categoria_mas_prestada = categoria

print("\nCategoría más prestada:", categoria_mas_prestada)

# Mostrar libros con más de 20 días
print("\nLibros con más de 20 días:")
for prestamo in prestamos:
    if prestamo["dias"] > 20:
        print(prestamo["titulo"], "-", prestamo["dias"], "días")
```

## Ejercicio 4: Procesador de datos

```python
def limpiar_datos(ventas):
    """Elimina valores negativos o None"""
    ventas_validas = []
    for v in ventas:
        if v is not None and v > 0:
            ventas_validas.append(v)
    return ventas_validas


def calcular_estadisticas(ventas):
    """Devuelve dict con total, promedio, max, min"""
    total = sum(ventas)
    promedio = total / len(ventas)
    maximo = max(ventas)
    minimo = min(ventas)

    return {
        "total": total,
        "promedio": promedio,
        "max": maximo,
        "min": minimo
    }


def clasificar_ventas(ventas, umbral_bajo, umbral_alto):
    """Devuelve dict con ventas bajas, medias, altas"""
    bajas = []
    medias = []
    altas = []

    for v in ventas:
        if v < umbral_bajo:
            bajas.append(v)
        elif v >= umbral_alto:
            altas.append(v)
        else:
            medias.append(v)

    return {
        "bajas": bajas,
        "medias": medias,
        "altas": altas
    }


def generar_reporte(ventas):
    """Usa las funciones anteriores y muestra reporte completo"""

    ventas_validas = limpiar_datos(ventas)
    eliminadas = len(ventas) - len(ventas_validas)

    estadisticas = calcular_estadisticas(ventas_validas)
    clasificacion = clasificar_ventas(ventas_validas, 2000, 4000)

    total_validas = len(ventas_validas)

    bajas = len(clasificacion["bajas"])
    medias = len(clasificacion["medias"])
    altas = len(clasificacion["altas"])

    print("=== REPORTE DE VENTAS ===\n")

    print(f"Datos procesados: {total_validas} ventas válidas ({eliminadas} inválidas eliminadas)\n")

    print("Estadísticas:")
    print(f"  Total: {estadisticas['total']}")
    print(f"  Promedio: {estadisticas['promedio']:.2f}")
    print(f"  Máximo: {estadisticas['max']}")
    print(f"  Mínimo: {estadisticas['min']}\n")

    print("Clasificación:")
    print(f"  Ventas bajas (< 2000): {bajas} ({bajas/total_validas*100:.1f}%)")
    print(f"  Ventas medias: {medias} ({medias/total_validas*100:.1f}%)")
    print(f"  Ventas altas (>= 4000): {altas} ({altas/total_validas*100:.1f}%)")


# Datos de prueba
ventas_brutas = [1500, -200, 2500, None, 3000, 0, 4500, -100, 2800, 5000]

generar_reporte(ventas_brutas)
```
